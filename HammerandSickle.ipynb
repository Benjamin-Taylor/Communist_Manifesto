##  - Top 100, 200, 300 frequency of words
##  - Word cloud displayed as the Hammer and Sickle
## - Word size proportional to frequency



# TOP 100 WORDS

```python
import numpy as np
from PIL import Image
from os import path
import matplotlib.pyplot as plt
import random
import pylab
from wordcloud import WordCloud, STOPWORDS

def grey_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    return "hsl(0, 0%%, %d%%)" % random.randint(60, 100)

%pylab inline
pylab.rcParams['figure.figsize'] = (20,12)

d = path.dirname("/home/system/MarxWordCLoud/")

mask = np.array(Image.open(path.join(d, "hambig2.jpg")))

text = open("/home/system/MarxWordCLoud/Manifesto.txt").read()

wc = WordCloud(max_words=100, mask=mask, stopwords=STOPWORDS, margin=5,
               random_state=1).generate(text)

default_colors = wc.to_array()
plt.title("Manifest der Kommunistischen Partei")
plt.imshow(wc.recolor(color_func=grey_color_func, random_state=4))
plt.axis("off")
plt.show()
```
